# Deep Learning Example-1 (The Simplest Neuron Network)

**Step#1: Import libraries**

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# check the tensorflow library version
print(tf.__version__)
```

**Step#2: Create Neural Network Model**

```python
# Create Nueral Netwrok model with one layer and one neuron.
# Definition of Terms:
# Sequestial - The successive layers are defined in layers, hence the Sequential class is used.
# Keras - It is high level API of tensorflow which makes te implementation of NN quite easy
# Dense - It is used to define layers of connected neurons. Here it reflects only one layer with one neuron.
# Units - Units here define the number of nuerons in that layer.
# Input_shape - This parameter takes the shape of input value. In this case, we have only one value of Xs array.

model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

**Step#3: Compile Neural Network Model**

```python
# Compile the NN model

# loss function measures the guess that how good or how bad it was and 
# then it feeds its result to Optimizer function which updates the weight and bias in order to make the guess more accurate.
model.compile(optimizer='sgd', 
              loss='mean_squared_error',
              metrics='accuracy')
```

**Step#4: Create Numpy Array Data**

```python
# make numpy array for data
Xs=np.array([-1,0,1,2,3,4], dtype=float)
Ys=np.array([-3,-1,1,3,5,7], dtype=float)
```

**Step#5: Fit Neural Network Model**

```python
# train the NN model (Epoch=500)
model.fit(Xs,Ys,epochs=500)
```

```python
# predict the model for unkown value
print(model.predict([7]))
```
